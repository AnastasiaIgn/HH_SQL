```python
# текст запроса
query_3_11 = """
WITH region_stats AS (
    SELECT 
        a.id AS area_id,
        a.name AS area_name,
        COUNT(DISTINCT e.id) AS employers_count,
        COUNT(DISTINCT v.id) AS vacancies_count
    FROM areas a
    LEFT JOIN employers e ON a.id = e.area  -- исправлено: e.area вместо e.area_id
    LEFT JOIN vacancies v ON a.id = v.area_id
    GROUP BY a.id, a.name
)
SELECT 
    area_name,
    employers_count,
    vacancies_count
FROM region_stats
WHERE vacancies_count = 0
ORDER BY employers_count DESC
LIMIT 1;
"""
```

```python
# результат запроса

df_result = pd.read_sql_query(query_3_11, connection)
print("Регион без вакансий с наибольшим количеством работодателей:")
print(df_result)
```

Для компании «Яндекс» выведите список [городов-миллионников](https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8), в которых представлены вакансии компании, вместе с количеством вакансий в этих регионах. Также добавьте строку "Total" с общим количеством вакансий компании. Результат отсортируйте по возрастанию количества.

    Если возникнут трудности с этим заданием, посмотрите материалы модуля 6.4 «Как получать данные из веб-источников и API».

```python
# код для получения списка городов-милионников

million_cities = [
    'Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург', 'Казань',
    'Красноярск', 'Нижний Новгород', 'Челябинск', 'Уфа', 'Краснодар',
    'Самара', 'Ростов-на-Дону', 'Омск', 'Воронеж', 'Пермь', 'Волгоград'
]

print(f"Список городов-миллионников: {len(million_cities)} городов")

cities_sql = ", ".join([f"'{city}'" for city in million_cities])

```

```python
# текст запроса
query_3_16 = f"""
WITH yandex_vacancies AS (
    -- Находим количество вакансий Яндекса по всем регионам
    SELECT 
        a.name AS city_name,
        COUNT(v.id) AS vacancy_count
    FROM vacancies v
    JOIN employers e ON v.employer_id = e.id
    JOIN areas a ON v.area_id = a.id
    WHERE e.name = 'Яндекс'  -- Используем точное совпадение
    GROUP BY a.id, a.name
),
filtered_cities AS (
    -- Основной запрос: выбираем только города-миллионники
    SELECT 
        city_name AS city,
        vacancy_count AS count
    FROM yandex_vacancies
    WHERE city_name IN ({cities_sql})
    
    UNION ALL
    
    -- Добавляем строку с итоговым количеством
    SELECT 
        'Total' AS city,
        SUM(vacancy_count) AS count
    FROM yandex_vacancies
    WHERE city_name IN ({cities_sql})
)
-- Внешний запрос для сортировки
SELECT 
    city AS "Город-миллионник",
    count AS "Количество вакансий"
FROM filtered_cities
ORDER BY 
    CASE WHEN city = 'Total' THEN 1 ELSE 0 END,
    count ASC,
    city ASC;
"""
```

```python
# результат запроса
df_result = pd.read_sql_query(query_3_16, connection)

print("=" * 60)
print("ВАКАНСИИ КОМПАНИИ «ЯНДЕКС» В ГОРОДАХ-МИЛЛИОННИКАХ")
print("=" * 60)
print(df_result.to_string(index=False))
print("=" * 60)
```

    + требуемый опыт — «Нет опыта» **или**

```python
# текст запроса
query_3_21 = """
WITH data_scientist_vacancies AS (
    -- Вакансии для Data Scientist
    SELECT 
        id,
        key_skills
    FROM vacancies
    WHERE (
        LOWER(name) LIKE '%data scientist%'
        OR LOWER(name) LIKE '%data science%'
        OR LOWER(name) LIKE '%исследователь данных%'
        OR (LOWER(name) LIKE '%ml%' AND LOWER(name) NOT LIKE '%html%')
        OR LOWER(name) LIKE '%machine learning%'
        OR LOWER(name) LIKE '%машинн%обучен%'
    )
    AND key_skills IS NOT NULL  -- Только вакансии с указанными навыками
),
skills_analysis AS (
    SELECT 
        -- Считаем количество навыков по количеству запятых + 1
        LENGTH(key_skills) - LENGTH(REPLACE(key_skills, ',', '')) + 1 AS skills_count
    FROM data_scientist_vacancies
)
SELECT 
    ROUND(AVG(skills_count)::numeric, 2) AS avg_skills_per_vacancy,
    MIN(skills_count) AS min_skills,
    MAX(skills_count) AS max_skills,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY skills_count) AS median_skills,
    COUNT(*) AS total_vacancies_with_skills
FROM skills_analysis;
"""
```

```python
# результат запроса

df_avg_skills = pd.read_sql_query(query_3_21, connection)
avg_skills = df_avg_skills['avg_skills_per_vacancy'].iloc[0]

print(f"Среднее количество ключевых навыков в DS-вакансиях: {avg_skills}")
print("\nПолная статистика:")
print(df_avg_skills)
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в среднем указывают для каждого типа требуемого опыта (уникальное значение из поля `experience`).

    При решении задачи примите во внимание следующее:
    1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
    2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
    3. Если в расчётах участвует `null`, в результате он тоже даст `null` (посмотрите, что возвращает запрос `select 1 + null`). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит `null` на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

    Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до **целого числа**.

```python
# текст запроса
query_3_22 = """
WITH data_scientist_vacancies AS (
    -- Вакансии для Data Scientist
    SELECT 
        id,
        experience,
        salary_from,
        salary_to
    FROM vacancies
    WHERE (
        LOWER(name) LIKE '%data scientist%'
        OR LOWER(name) LIKE '%data science%'
        OR LOWER(name) LIKE '%исследователь данных%'
        OR (LOWER(name) LIKE '%ml%' AND LOWER(name) NOT LIKE '%html%')
        OR LOWER(name) LIKE '%machine learning%'
        OR LOWER(name) LIKE '%машинн%обучен%'
    )
    AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)  -- Только с зарплатой
),
calculated_salaries AS (
    SELECT 
        experience,
        -- Расчет зарплаты: если оба поля заполнены - среднее, иначе заполненное поле
        CASE 
            WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL 
                THEN (COALESCE(salary_from, 0) + COALESCE(salary_to, 0)) / 2.0
            WHEN salary_from IS NOT NULL THEN COALESCE(salary_from, 0)::float
            WHEN salary_to IS NOT NULL THEN COALESCE(salary_to, 0)::float
        END AS calculated_salary
    FROM data_scientist_vacancies
    WHERE (salary_from IS NOT NULL OR salary_to IS NOT NULL)
)
SELECT 
    experience,
    COUNT(*) AS vacancy_count,
    ROUND(AVG(calculated_salary)) AS avg_salary
FROM calculated_salaries
GROUP BY experience
ORDER BY 
    CASE experience
        WHEN 'Нет опыта' THEN 1
        WHEN 'От 1 года до 3 лет' THEN 2
        WHEN 'От 3 до 6 лет' THEN 3
        WHEN 'Более 6 лет' THEN 4
        ELSE 5
    END;
"""
```

```python
# результат запроса
df_salaries = pd.read_sql_query(query_3_22, connection)
print("Средняя зарплата Data Scientist по уровням опыта:")
print(df_salaries)
```
